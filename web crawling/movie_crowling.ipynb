```python
def getReview(page_index):

    print("*"*30)
    url = 'https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=167635&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page=' + str(page_index)
    
    with urllib.request.urlopen(url) as response:
        html = response.read()
        soup = BeautifulSoup(html, 'html.parser')
        score_result = soup.find('div', class_ = 'score_result')
        rows = score_result.find_all('li')
        
        print("loading html completed")
        print("page : " + str(page_index))
        return [[
                float(row.find('span', class_ = 'st_on')['style'].split(':')[1][:-1]), 
                row.find('div', class_ = 'score_reple').find_all('span')[0].text.strip(' \t\n\r') if len(row.find('div', class_ = 'score_reple').find_all('span')) < 3 else row.find('div', class_ = 'score_reple').find_all('span')[1].text.strip(' \t\n\r'),
                row.find_all('em')[2].text
            ] for row in rows]
   
```

```python
result = []
temp = []
for page_index in count():
    data = getReview(page_index+1)
    if temp != data :
        temp = data
        result += data
    else :
        print("종료")
        break;
```
