# Model 1

```python
# 필요한 라이브러리
import numpy as np
import tensorflow as tf
tf.__version__
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# 실행할 때마다 같은 결과를 출력하기 위한 seed 값 설정
seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# 준비된 수술 환자 데이터를 불러들임
Data_set = np.loadtxt("dataset/ThoraricSurgery.csv", delimiter=",")
Data_set
```

```python
# 환자의 기록과 수술 결과를 X와 Y로 구분하여 저장
# x: input, Y: result
X = Data_set[:, 0:17]
Y = Data_set[:, 17]
```

```python
# 딥러닝 모델 설정
model = Sequential([
    Dense(30, input_shape=(17,), activation='relu'),
    Dense(1, activation='sigmoid')
])
model.summary()
```

```python
# 딥러닝 환경 설정 (오차함수, 최적화함수)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# 딥러닝 실행
model.fit(X, Y, epochs=20, batch_size=10)
```

```python
# 결과 출력
# evaluate() returns the loss value & metrics values for the model 
print("\n Accuracy: %.4f" % (model.evaluate(X, Y, verbose=2))[1])
```

# Model 2

```python
# 딥러닝 모델 설정
model2 = Sequential([
    Dense(33, input_shape=(17,), activation='relu'),
    Dense(11, activation='relu'),
    Dense(1, activation='sigmoid')
])
model2.summary()
```

```python
# 딥러닝 환경 설정 (오차함수, 최적화함수)
model2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# 딥러닝 실행
model2.fit(X, Y, epochs=500, batch_size=10)
```

```python
# 결과 출력
# evaluate() returns the loss value & metrics values for the model 
print("\n Accuracy: %.4f" % (model2.evaluate(X, Y, verbose=2))[1])
```
